### Î¨∏Ï†ú Ìï¥ÏÑù
- ÏûÖÎ†•Í∞í: Î¨∏ÏûêÏó¥ products, searchWord
> `searchWord`Ïùò Í∞Å Îã®Ïñ¥Î•º ÏûëÏÑ±Ìïú Îí§, `products`ÏóêÏÑú ÏµúÎåÄ 3Í∞úÏùò Ïù¥Î¶ÑÏùÑ Ï†úÏïàÌï¥Îùº. <br/>
Ï†úÏïàÎêú productsÎäî `searchWord`ÎùºÎäî Í≥µÌÜµ Ï†ëÎëêÏÇ¨Î•º Í∞ÄÏ†∏Ïïº ÌïòÍ≥†, Ïù¥ Í≥µÌÜµ Ï†ëÎëêÏÇ¨Í∞Ä 3Í∞ú Ïù¥ÏÉÅÏù¥ÎùºÎ©¥ 3Í∞úÏùò lexicographically minimum productÎ•º Î∞òÌôòÌï¥Îùº
- Î∞òÌôòÍ∞í: `searchWord`Ïùò Í∞ÅÍ∞ÅÏùò Îã®Ïñ¥Í∞Ä ÏûëÏÑ±Îêú Ïù¥ÌõÑ Ï†úÏïàÎêú productsÏùò Î¶¨Ïä§Ìä∏ Î∞òÌôòÌïòÍ∏∞

> ‚ùì `lexicographically minimum product`
> - `lexicographically`: ÏÇ¨Ï†Ñ ÏàúÏÑúÎåÄÎ°ú
> - ÏÇ¨Ï†Ñ ÏàúÏÑúÎåÄÎ°ú ÏïûÏóêÏÑúÎ∂ÄÌÑ∞ 3Í∞úÏùò product Î∞òÌôòÌïòÍ∏∞

```python
class Solution(object):
    def suggestedProducts(self, products, searchWord):
        
        products = sorted(products) # ÏÇ¨Ï†Ñ Ïàú Ï†ïÎ†¨
        
        result = [[] for i in range(len(searchWord))] # Í∏∏Ïù¥Í∞Ä Ï†ïÌï¥ÏßÑ Î¶¨Ïä§Ìä∏ ÎßåÎì§Í∏∞ (searchWord Í∏∏Ïù¥ÎßåÌÅº Î¶¨Ïä§Ìä∏ ÏÉùÍπÄ)
        
        for i in range(len(searchWord)):
            for product in products:
                if (searchWord[0:i+1] == product[0:i+1]) and (len(result[i]) < 3):
                    result[i].append(product)

        return result
```

```python
sol = Solution()
sol.suggestedProducts(products=['mobile', 'mouse', 'moneypot', 'monitor', 'mousepad'], searchWord='mouse')
```
