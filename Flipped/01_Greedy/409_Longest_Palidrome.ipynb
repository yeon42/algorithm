```python
from collections import Counter
```

```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: int
        """
        answer = 0
        
        # Í∞Å ÏõêÏÜåÍ∞Ä ÏßùÏàò Í∞úÎ°ú ÏûàÏñ¥Ïïº ÌöåÎ¨∏Ïóê Ïì∏ Ïàò ÏûàÏùå
        counter = Counter(s)
        odd = []
        
        for c in counter.values():
            if c % 2 == 0:
                answer += c # ÏßùÏàò Í∞úÎ©¥ Í∑∏ Í∏∏Ïù¥ÎßåÌÅº ÌöåÎ¨∏Ïóê Îì§Ïñ¥Í∞à Ïàò ÏûàÏùå
            else:
                odd.append(c)
        
        if len(odd) > 0: # ÌôÄÏàò Í∞úÏùò ÏïåÌååÎ≤≥Ïù¥ ÏûàÎã§Î©¥
            answer += max(odd) # Í∑∏ ÏµúÎåìÍ∞í
                
        return answer
```
